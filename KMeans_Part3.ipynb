```python
from sklearn import cluster

def quantize(image, n_colors):
    width, height, depth = image.shape
    reshaped_image = np.reshape(image, (width * height, depth))

    model = cluster.KMeans(n_clusters = n_colors)
    labels = model.fit_predict(reshaped_image)
    palette = model.cluster_centers_

    quantized_image = np.reshape(palette[labels], (width, height, palette.shape[1]))

    return quantized_image
```

```python
image1_re_4 = quantize(image1, 4)
```

```python
image1_re_16 = quantize(image1, 16)
```

```python
image1_re_8 = quantize(image1, 8)
```
