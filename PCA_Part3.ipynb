```python
import imageio
import requests
from io import BytesIO
import numpy as np

url = 'http://www.utdallas.edu/~axn112530/cs6375/unsupervised/images/image1.jpg'
response = requests.get(url)
image1 = imageio.imread(BytesIO(response.content))

url = 'http://www.utdallas.edu/~axn112530/cs6375/unsupervised/images/image2.jpg'
response = requests.get(url)
image2 = imageio.imread(BytesIO(response.content))

url = 'http://www.utdallas.edu/~axn112530/cs6375/unsupervised/images/image3.jpg'
response = requests.get(url)
image3 = imageio.imread(BytesIO(response.content))
```

```python
from sklearn.decomposition import PCA

def compress_pca(image, N = 1):
    width, height, depth = image.shape
    reshaped_image = np.reshape(image, (-1, 3))
    pca = PCA(N)
    lower_dimensional_data = pca.fit_transform(reshaped_image)
    approximation = pca.inverse_transform(lower_dimensional_data)
    appx = np.reshape(approximation, (width, height, depth))
    return appx
```

```python
apx11 = compress_pca(image1, N = 1)
apx12 = compress_pca(image1, N = 2)
apx13 = compress_pca(image1, N = 3)
```

```python
apx21 = compress_pca(image2, N = 1)
apx22 = compress_pca(image2, N = 2)
apx23 = compress_pca(image2, N = 3)
```

```python
apx31 = compress_pca(image3, N = 1)
apx32 = compress_pca(image3, N = 2)
apx33 = compress_pca(image3, N = 3)
```
